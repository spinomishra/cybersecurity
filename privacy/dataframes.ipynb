```python
def scanforfiles(torque_dir):
        torque_files = []
        dirscan = os.scandir(torque_dir)
        for item in dirscan:
            try:
                if item.is_dir():
                    try:
                        torque_files.extend(scanforfiles(item.path))
                    except:
                        print("skipping one folder due to exception")
                elif item.is_file():
                    if item.name.endswith(".csv"):
                        torque_files.append(item.path)
            except:
                print("skipping one item due to exception")

        dirscan.close()
        return torque_files
```

#Privacy metrics

```python
def k_Anonymity(data, qid):
    data_temp = data
    for q in qid:
      data_temp = data_temp.sort_values(q, ignore_index=True)


    # # Get quasi_identifiers data only
    # qi_data = data[qid]

    # Get list of EC (equivalence classes) and associated counts for each EC
    equivalece_classes = data_temp.groupby(qid).size().reset_index().rename(columns={0:'count'})

    # Get lowest EC class count = k
    k = min(equivalece_classes['count'])
    return k, equivalece_classes
```

```python
k,eq = k_Anonymity(df, quasiId)
print ("K Anonimity of raw data is ", k)
display(eq)
```
